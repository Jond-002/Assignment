```python
# Assignment: Asymptotic Notations and Time Complexity

## (a) Definitions

### i. Big-O Notation

Big-O notation shows the maximum time an algorithm can take to run. It helps us understand the worst-case performance when the input size becomes large.

---

### ii. Big-Θ Notation

Big-Theta notation shows the exact running time of an algorithm. It gives both the upper and lower limits of the function.

---

### iii. Big-Ω Notation

Big-Omega notation shows the minimum time an algorithm can take. It represents the best-case performance.
```

```python
### i. Time Complexity

The first loop runs n times and the second loop also runs n times.

This means the statement inside runs n × n times.

So, the time complexity is:

O(n²)
```

```python
### ii. Efficiency

This algorithm is not efficient for large values of n.

This is because the number of operations increases very fast when n increases.

For example, if n is 1000, the loop will run 1,000,000 times.

This makes the program slow when the input is large.
```

```python
# 3 (a)
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # Return index if found
    return -1  # Return -1 if not found


arr = [21, 10, 45, 32, 67, 5, 18]
result = linear_search(arr, 32)

print("Element found at index:", result)
```

```python
# 3 (b)
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1


arr = [21, 10, 45, 32, 67, 5, 18]
arr.sort()  # Sorting before binary search
result = binary_search(arr, 45)

print("Sorted list:", arr)
print("Element found at index:", result)
```

```python
# 3 (c)
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr


arr = [21, 10, 45, 32, 67, 5, 18]
sorted_arr = bubble_sort(arr)

print("Sorted list:", sorted_arr)
```

```python
# 4 (a)
class Node:
    def __init__(self, data):
        self.data = data      # Store data
        self.next = None      # Pointer to next node
```

```python
# 4 (b)
class LinkedList:
    def __init__(self):
        self.head = None   # Initially the list is empty

    # i. Insert at the beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
```

```python
 # ii. Insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:   # If list is empty
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
```

```python
 # iii. Display all elements
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

```python
# 5 (b)
class Queue:
    def __init__(self):
        self.queue = []

    # i. Add a student to the queue (Enqueue)
    def add_student(self, name):
        self.queue.append(name)
        print(name, "has joined the queue.")
```

```python
 # ii. Serve (remove) the first student (Dequeue)
    def serve_student(self):
        if len(self.queue) == 0:
            print("No students in the queue.")
        else:
            served = self.queue.pop(0)
            print(served, "has been served.")
```

```python
 # iii. Display the current queue
    def display_queue(self):
        if len(self.queue) == 0:
            print("The queue is empty.")
        else:
            print("Current Queue:")
            for student in self.queue:
                print(student)
```
